```python
import pandas as pd

data_sampah = pd.read_csv('data_sampah.csv')

total_sampah_pertahun = {}

for index, row in data_sampah.iterrows():
    tahun = row['tahun']
    produksi = row['jumlah_produksi_sampah']
    
    if tahun in total_sampah_pertahun:
        total_sampah_pertahun[tahun] += produksi
    else:
        total_sampah_pertahun[tahun] = produksi

for tahun, total in total_sampah_pertahun.items():
    print(f'Total produksi sampah di tahun {tahun} adalah: {total:.2f} ton')
```

```python
import pandas as pd

data_sampah = pd.read_csv('data_sampah.csv')

total_produksi_perkota_tahun = {}

for index, row in data_sampah.iterrows():
    kota = row['nama_kabupaten_kota'] 
    tahun = row['tahun']                    
    produksi = row['jumlah_produksi_sampah']      
    
    if kota not in total_produksi_perkota_tahun:
        total_produksi_perkota_tahun[kota] = {}
    
    if tahun in total_produksi_perkota_tahun[kota]:
        total_produksi_perkota_tahun[kota][tahun] += produksi
    else:
        total_produksi_perkota_tahun[kota][tahun] = produksi

for kota_kabupaten, tahun_data in total_produksi_perkota_tahun.items():
    for tahun, total in tahun_data.items():
        print(f'Total produksi sampah di {kota_kabupaten} pada tahun {tahun} adalah: {total:.2f} ton')
```

```python
import pandas as pd

data_sampah = pd.read_csv('data_sampah.csv')

tahun_produksi = 2015 

total_sampah_tahun_tertentu = 0

for index, row in data_sampah.iterrows():
    if row['tahun'] == tahun_produksi:
        total_sampah_tahun_tertentu += row['jumlah_produksi_sampah']

print(f'Total produksi sampah di seluruh Kabupaten/Kota di Jawa Barat pada tahun {tahun_produksi} adalah: {total_sampah_tahun_tertentu:.2f} ton')

hasil_tahun_tertentu = {
    'Tahun': [tahun_produksi],
    'Total Produksi Sampah (ton)': [total_sampah_tahun_tertentu]
}

df_hasil_tahun_tertentu = pd.DataFrame(hasil)

df_hasil_tahun_tertentu.to_csv('total_produksi_sampah_2015.csv', index=False)

df_hasil_tahun_tertentu.to_excel('total_produksi_sampah_2015.xlsx', index=False)

print("Hasil telah diekspor ke 'total_produksi_sampah_2015.csv' dan 'total_produksi_sampah_2015.xlsx'.")
```

```python
import pandas as pd

data_sampah = pd.read_csv('data_sampah.csv')

total_sampah_pertahun = {}

for index, row in data_sampah.iterrows():
    tahun = row['tahun']  
    produksi = row['jumlah_produksi_sampah']  
    
    if tahun in total_sampah_pertahun:
        total_sampah_pertahun[tahun] += produksi
    else:
        total_sampah_pertahun[tahun] = produksi

for tahun, total in total_sampah_pertahun.items():
    print(f'Total produksi sampah di tahun {tahun} adalah: {total:.2f} ton')

hasil_pertahun = {
    'Tahun': list(total_sampah_pertahun.keys()),
    'Total Produksi Sampah (ton)': list(total_sampah_pertahun.values())
}

df_hasil_pertahun = pd.DataFrame(hasil)

df_hasil_pertahun.to_csv('total_produksi_sampah_per_tahun.csv', index=False)

df_hasil_pertahun.to_excel('total_produksi_sampah_per_tahun.xlsx', index=False)

print("Hasil telah diekspor ke 'total_produksi_sampah_per_tahun.csv' dan 'total_produksi_sampah_per_tahun.xlsx'.")
```

```python
import pandas as pd

data_sampah = pd.read_csv('data_sampah.csv')

total_produksi_perkota_tahun = {}

for index, row in data_sampah.iterrows():
    kota = row['nama_kabupaten_kota']  
    tahun = row['tahun']                
    produksi = row['jumlah_produksi_sampah']  
    
    if kota not in total_produksi_perkota_tahun:
        total_produksi_perkota_tahun[kota] = {}
    
    if tahun in total_produksi_perkota_tahun[kota]:
        total_produksi_perkota_tahun[kota][tahun] += produksi
    else:
        total_produksi_perkota_tahun[kota][tahun] = produksi

for kota_kabupaten, tahun_data in total_produksi_perkota_tahun.items():
    for tahun, total in tahun_data.items():
        print(f'Total produksi sampah di {kota_kabupaten} pada tahun {tahun} adalah: {total:.2f} ton')

hasil_perkota_pertahun = {
    'Kota/Kabupaten': [],
    'Tahun': [],
    'Total Produksi Sampah (ton)': []
}

for kota_kabupaten, tahun_data in total_produksi_perkota_tahun.items():
    for tahun, total in tahun_data.items():
        hasil_perkota_pertahun['Kota/Kabupaten'].append(kota_kabupaten)
        hasil_perkota_pertahun['Tahun'].append(tahun)
        hasil_perkota_pertahun['Total Produksi Sampah (ton)'].append(total)

df_hasil_perkota_pertahun = pd.DataFrame(hasil)

df_hasil_perkota_pertahun.to_csv('total_produksi_sampah_perkota_tahun.csv', index=False)

df_hasil_perkota_pertahun.to_excel('total_produksi_sampah_perkota_tahun.xlsx', index=False)

print("Hasil telah diekspor ke 'total_produksi_sampah_per_kota_tahun.csv' dan 'total_produksi_sampah_per_kota_tahun.xlsx'.")
```
